```python
plt.hist(df['tot_amt_v_depend'])
```

```python
df[['n_depends', 'total_amount']].corr()
```

- More than 800 people spend around 100 bucks per dependent

```python
df[['n_depends', 'total_purchase']].corr()
```

- No. of dependents and total no. of purchases are negatively correlated
- Most people purchase around 8 items per dependent
